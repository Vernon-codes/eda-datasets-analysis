```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
df_train=pd.read_excel(r"D:\Development\EDA\Practice\Raw\flight_train.xlsx")
df_train.head()
```

```python
df_test=pd.read_excel(r"D:\Development\EDA\Practice\Raw\flight_test.xlsx")
df_test.head()
```

```python
#merging both train and test data
final_df=pd.concat([df_train,df_test])
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Day']=final_df['Date_of_Journey'].str.split('/').str[0].astype(int)
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1].astype(int)
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2].astype(int)
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.head()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].str.split(' ').str[0]
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head()
```

```python
final_df.isnull().sum()

final_df['Total_Stops']=final_df['Total_Stops'].fillna(final_df['Total_Stops'].mode()[0])
final_df['Route']=final_df['Route'].fillna(final_df['Route'].mode()[0])

final_df.isnull().sum()
final_df.info()
```

```python
final_df['Arrival_min']=final_df['Arrival_Hour'].str.split(':').str[1].astype(int)
final_df['Arrival_Hour']=final_df['Arrival_Hour'].str.split(':').str[0].astype(int) 

final_df.head()
```

```python
final_df.info()
```

```python
final_df['Dep_Time_min']=final_df['Dep_Time'].str.split(':').str[1].astype(int)
final_df['Dep_Time_hour']=final_df['Dep_Time'].str.split(':').str[0].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)

final_df.head()
```

```python
final_df.info()
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
final_df.head()
```

```python
final_df.info(verbose=True)
```

```python
final_df.head()
#convert duration into hours and minutes and combine them as total minutes
```

```python
final_df['Hours']=final_df['Duration'].str.split(' ').str[0].str.replace('h','')
final_df['Minutes']=final_df['Duration'].str.split(' ').str[1].str.replace('m','') 
final_df['Hours']=pd.to_numeric(final_df['Hours'],errors='coerce') 
final_df['Minutes']=pd.to_numeric(final_df['Minutes'],errors='coerce') 
final_df['Total_Duration_min']=final_df['Hours']*60+final_df['Minutes']
final_df.drop(['Duration','Hours','Minutes','Total_Duration_min'],axis=1,inplace=True) 
# because in some cases duration is only in hours or minutes and some it only takes 5 mins which is not possible
final_df.head()
```

```python
final_df.info()
final_df['Airline'].unique()
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
final_df['Airline']=le.fit_transform(final_df['Airline'])
final_df['Source']=le.fit_transform(final_df['Source'])
final_df['Additional_Info']=le.fit_transform(final_df['Additional_Info'])
final_df['Destination']=le.fit_transform(final_df['Destination'])
final_df.head()
```

```python
final_df.to_excel(r"D:\Development\EDA\Practice\cleaned\flight_preprocessed.xlsx",index=False)
```
