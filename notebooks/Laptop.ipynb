# Laptop Dataset 
# Note: Due to some errors I Have removed the Laptop Ram Column

```python
#import all necessary libraries
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read the dataset
df=pd.read_csv("D:\Development\EDA\Practice\Raw\laptopData.csv")
df.drop(columns=['Unnamed: 0'],inplace=True) 
df.head()
```

```python
#Getting the columns of the dataset
df.columns
```

```python
df.info()
```

```python
df.shape
```

```python
#Handling missing values

df.isnull().sum() #Number of Missing values per column
df.isnull().sum()/len(df)*100 #Percentage of missing values
```

```python
#1.Company Column
df['Company']=df['Company'].fillna(df['Company'].mode()[0])
df.info()

#2.Type Name Column
df['TypeName']=df['TypeName'].fillna(df['TypeName'].mode()[0])
df.info()

#3.Inches Column
df['Inches']=pd.to_numeric(df['Inches'],errors='coerce') #Converting to numeric  
df['Inches']=df['Inches'].fillna(df['Inches'].mean()) #Numerical column so we use mean
df.info()

#4.Screen Resolution Column
df['ScreenResolution']=df['ScreenResolution'].fillna(df['ScreenResolution'].mode()[0])
df.info()

#5.CPU Column
df['Cpu']=df['Cpu'].fillna(df['Cpu'].mode()[0])
df.info()

#6.Ram Column
df.drop(columns=['Ram'],inplace=True,axis=1) #Dropping the column as it has more than 40% missing values
df.info()
```

```python

#8 GPU Colummn
df['Gpu']=df['Gpu'].fillna(df['Gpu'].mode()[0])
df.info()

#9.Operating System Column
df['OpSys']=df['OpSys'].fillna(df['OpSys'].mode()[0])
df.info()

#10.Weight Column
df['Weight']=df['Weight'].astype(str).str.replace('kg','') #Removing 'kgs' if any
df['Weight']=pd.to_numeric(df['Weight'],errors='coerce') #Converting to numeric
df['Weight']=df['Weight'].fillna(df['Weight'].mean()) #Numerical column so we use mean
df.info()   


# 11.Price Column
df['Price']=df['Price'].fillna(df['Price'].mean()) #Numerical column so we use mean
df.info()
```

```python
#Separating the screen resolution column into multiple columns
df['Resolution']=df['ScreenResolution'].str.extract(r'(\d+x\d+)')
df[['width','height']]=df['Resolution'].str.split('x',expand=True)
```

```python
df['Cpu'].unique()
#Extract CPU Name (removing clock speed)
df['CpuName'] = df['Cpu'].str.replace(r'\s*\d+(\.\d+)?GHz', '', regex=True).str.strip()

# Extract Clock Speed in GHz (only the number)
df['CpuClockSpeed']= df['Cpu'].str.extract(r'(\d+(\.\d+)?)GHz')[0]

# Convert clock speed to numeric (float)
df['CpuClockSpeed'] = pd.to_numeric(df['CpuClockSpeed'], errors='coerce')
```

```python
df['Memory_space']=df['Memory'].str.split().str[0] #Extracting the memory space
df.drop(columns=['Memory'],inplace=True,axis=1) #Dropping the original Memory column
df['Memory_space']=df['Memory_space'].str.replace('GB','',regex=True) #Removing GB and TB
df['Memory_space']=df['Memory_space'].str.replace('TB','',regex=True) #Removing GB and TB
```

```python
df.head()
```

```python
df['Memory_space']=df['Memory_space'].fillna(df['Memory_space'].mode()[0])
df['Memory_space']=pd.to_numeric(df['Memory_space'],errors='coerce') #Converting to numeric
df.info()
```

```python
df['Gpu_brand']=df['Gpu'].str.split().str[0] #Extracting GPU brand
df['Gpu_model']=df['Gpu'].str.extract(r'(\d+)$')#Exracting GPU model number 
df['Cpu_brand']=df['CpuName'].str.split().str[0] #Extracting CPU brand

df['Gpu_brand'].head()
df['Gpu_model'].head()
df['Cpu_brand'].head()
# df['CpuName'].unique()
```

```python
df['Cpu_series'] = df['CpuName'].str.extract(r'(Core i\d|Core M [a-z0-9]+|Pentium|Atom|Celeron|Ryzen \d|A\d-Series|FX \d+|Cortex [A-Z0-9&]+)')
df['Cpu_model_number'] = df['CpuName'].str.extract(r'(\d{3,4}[A-Z]?\d?|\d{3,4}[A-Z]?)')
df['Cpu_series'].head()
df['Cpu_model_number']=df['Cpu_model_number'].fillna(df['Cpu_model_number'].mode()[0])
```

```python
df.drop(columns=['ScreenResolution','Resolution','Cpu','Gpu','CpuName'],inplace=True,axis=1)
```

```python
df.head()
```

```python
df['Gpu_model']=pd.to_numeric(df['Gpu_model'],errors='coerce')
df['Gpu_model'].fillna(df['Gpu_model'].mean(),inplace=True)
df['Gpu_model'].isnull().sum()

df['Cpu_series'].fillna('unknown',inplace=True)
# df['Cpu_series']=pd.to_numeric(df['Cpu_series'],errors='coerce')
df['Cpu_series'].isnull().sum()
df['Cpu_series'].unique()
```

```python
df.info()
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df.info()
df.shape
```

```python
df['Memory_space'].fillna(df['Memory_space'].mean(),inplace=True)
df.info()
```
